Программист Миша владеет биткоинами, кошелек с которыми хранит на флешке с аппаратным шифрованием. Пароль от нее он записал на бумажке, которую потерял, а самостоятельно вспомнить нужную комбинацию ему не под силу. Единственное, что он знает про свой пароль, это что он является счастливым числом - то есть сумма первой половины числа совпадает со второй половиной числа, и он делится на 3, 13 и 73. Например, число 12344321 - счастливое. Известно, что пароль был шестизначным. Определите все возможные пароли, которые соответствуют условию.

```python
def is_digit_happy(n):
  a = list(map(int, list(str(n))))  # Преобразуем число в строку, затем в список символов. Далее функцией map преобразуем каждый элемент списка в число
  if sum(a[:4]) == sum(a[4:]):      # Сравниваем суммы первой половины и второй
    return True
  return False


for i in range(100000, 1000000):
  if is_digit_happy(i) and i % 73 == 0 and i % 3 == 0 and i % 13 == 0:  # проверяем все условия
    print(i)
```

Ольга Бузова не остановилась на создании ресторана и Бузкоина и решила воплотить свои новую безумную идею - БузБанк. Дизайн пластиковых карт был давно разработан, осталось придумать красивые номера. Номер карты состоит из 10 цифр. Он уникален для конкретного банка и определенной платежной системы. Известно, что первые и последние четыре цифры 1986. Найти все номера карт БузБанка, которые при вышесказанных условиях делятся на 3 и 7

```python
a1 = 1986001986  # Выбираем диапазон. Логично, что число имеет слудующую маску 1986--1986. Минимальное 1986001986 и максимальное 1986991986
a2 = 1986991986

for i in range(a1, a2 + 1):                                   # не забываем про то, что в питоне правая граница в range не включается, поэтому добавляем единичку
  if str(i)[-4:] == '1986' and i % 3 == 0 and i % 7 == 0:     # str(i)[-4:] разберем это выражение. Функция str(i) 
                                                              # преобразует число i в строку. Срез [-4:] берет последние 4 символа
    print(i)
```
